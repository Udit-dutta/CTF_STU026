```python
import pandas as pd
import numpy as np
import hashlib
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
# import shap
import nltk
```

```python
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

TARGET_TITLE = "The Ice Child"     
YOUR_STUDENT_ID = '026'           
HASH_KEY = '927F09DE'              
SHORT_LIMIT = 15                  
SUPERLATIVES = ['best', 'greatest', 'amazing', 'perfect', 'superb', 'wonderful', 'excellent', 'love', 'awesome']

print("Setup complete.")
```

```python
# Load the files
books_df = pd.read_csv('books.csv')
reviews_df = pd.read_csv('reviews.csv')
target_books = books_df[books_df['title'].str.strip() == TARGET_TITLE]
book_ids = target_books['parent_asin'].unique()

print(f"Found Book IDs: {book_ids}")

book_reviews = reviews_df[
    (reviews_df['parent_asin'].isin(book_ids)) &
    (reviews_df['rating'] == 5.0)
].copy()

book_reviews = book_reviews[
    ~book_reviews['text'].astype(str).str.contains(HASH_KEY, case=False, na=False)
]

print(f" collected {len(book_reviews)} genuine reviews for training.")
```

```python
# 1. Create Features
book_reviews['word_count'] = book_reviews['text'].astype(str).apply(lambda x: len(x.split()))
book_reviews['has_superlative'] = book_reviews['text'].astype(str).str.lower().apply(
    lambda x: any(word in x for word in SUPERLATIVES)
)

book_reviews['suspicion_label'] = np.where(
    (book_reviews['word_count'] <= SHORT_LIMIT) & 
    (book_reviews['has_superlative']),
    1, 0
)

vectorizer = CountVectorizer(stop_words='english', min_df=1)

X = vectorizer.fit_transform(book_reviews['text'].astype(str)).toarray()
y = book_reviews['suspicion_label']

# 4. Train the Model
model = LogisticRegression(random_state=42)
model.fit(X, y)

print("Model trained successfully.")
```

```python
import shap
genuine_mask = (book_reviews['suspicion_label'] == 0)
X_genuine = X[genuine_mask]

explainer = shap.LinearExplainer(model, X)
shap_values = explainer.shap_values(X_genuine)

vals = shap_values[1] if isinstance(shap_values, list) else shap_values

mean_shap = np.mean(vals, axis=0)
feature_names = vectorizer.get_feature_names_out()

sorted_indices = np.argsort(mean_shap)

top_words = []
for i in sorted_indices:
    word = feature_names[i]
    if word.isalpha() and len(word) > 2: 
        top_words.append(word)
    if len(top_words) >= 3:
        break

print(f"Top 3 Genuine Words: {top_words}")

flag3_input = "".join(top_words) + YOUR_STUDENT_ID
flag3_hash = hashlib.sha256(flag3_input.encode('utf-8')).hexdigest()[:10]

print(f"\n FLAG3: FLAG3{{{flag3_hash}}}")
```
